```python
import numpy as np
```

##### defining helper function

```python
def sigmoid_numpy(X):       # sigmoid activation
    return 1/(1+np.exp(-X))
```

```python
def logloss(y_true, y_predicted):
    eps= 1e-10   # to get rid of error of exict
    y_predicted_new= [max(i, eps) for i in y_predicted]  #handling o value of predict
    y_predicted_new= [min(i, 1-i) for i in y_predicted_new]  #1's of pred
    y_predicted_new= np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*(1-y_predicted_new))
```

```python
def gradient_descent(age, affordibility, y_true, epochs):
    w1 = w2 = 1
    bias = 0
    lr= 0.5
    n = len(age)

    for i in range(epochs):
        weighted_sum= w1*age + w2*affordibility + bias
        y_predicted = sigmoid_numpy(weighted_sum)
        
        loss = logloss(y_true, y_predicted)
    
        # computed derivatives and update weights
        w1d = (1/n)*np.dot(np.transpose(age),(y_predicted-y_true))
        w2d = (1/n)*np.dot(np.transpose(affordibility),(y_predicted-y_true)) 

        bias_d = np.mean(y_predicted-y_true)
    
        w1 = w1 - w1d*lr
        w2 = w2 - w2d*lr
        bias = bias - bias_d*lr
    
        print(f"Epoch: {i}, w1: {w1}, w2: {w2}, bias: {bias}, w1d: {w1d}, w2d: {w2d}, loss: {loss}")
        
    return w1, w2, bias
```

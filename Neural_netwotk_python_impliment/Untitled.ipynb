### Neural Network Implemention in python

#### import required packages

```python
import numpy as np
import pandas as pd
```

#### Creating custom neural network

```python
class myNN:
    # it's task is to come up with best parameters including weights and bias
    
    def __init__(self):
        self.w1 = 1
        self.w2 = 1
        self.bias = 0
        
    def sigmoid_numpy(self, X):
        return 1/(1 + np.exp(-X))
    
    def log_loss(self, y_true, y_predicted):
        eps = 1e-15
        y_predicted_new = [max(i, eps) for i in y_predicted]
        y_predicted_new = [min(i, 1-i) for i in y_predicted_new]
        y_predicted_new = np.array(y_predicted_new)
        return -np.mean(y_true*np.log(y_predicted_new) + (1-y_true)*np.log(1-y_predicted_new))
        
    def gradient_descent(self, age, affordibility, y_true, epochs, loss_threshold= 0.03, lr= 0.5):
        w1 = w2 = 1
        bias = 0
        
        lr = lr
        m = len(age)
        
        # compute derivative and updates parameters until reaching desired loss or enough epochs
        
        for i in range(epochs):
            
            weighted_sum = w1*age + w2*affordibility + bias
            y_predicted = self.sigmoid_numpy(weighted_sum)
            
            loss= self.log_loss(y_true, y_predicted)
            
            w1d = (1/m) * np.dot(age,(y_predicted - y_true))
            w2d = (1/m) * np.dot(affordibility,(y_predicted - y_true))
            bias_d = np.mean(y_predicted - y_true)
            
            w1 = w1 - lr*w1d
            w2 = w2 - lr*w2d
            bias = bias - lr*bias_d
            
            print(f"epoch: {i}, w1:{w1}, w2: {w2}, bias: {bias}, loss: {loss}")
        
            
        return w1, w2, bias
    
    def fit(self, X, y, epochs= 5000):
        self.w1, self.w2, self.bias = self.gradient_descent(X["age"], X["affordibility"], y, epochs)
    
    def predict(self, X_test):
        weighted_sum = self.w1 * X_test["age"] + self.w2 * X_test["affordibility"] + self.bias
        return self.sigmoid_numpy(weighted_sum)   
```

#### Testing neural network

```python
import math
import matplotlib.pyplot as plt
```

```python
class Activations:
    def __init__(self, x):
        self.x = x
        
    def sigmoid(x):
        return 1/(1+math.exp(-x))
    
    def tanh(x):
        return (math.exp(x) - math.exp(-x))/(math.exp(x) + math.exp(-x))
    
    def ReLU(x):
        return max(0, x)
    
    def leakyReLU(x):
        return max(0.1*x, x)
    
# can convert x to self.x by calling self in function and passing self.x in corresponding expression
```

```python
compute = Activations
```

```python
compute.sigmoid(100)
```

```python
compute.sigmoid(1)
```

```python
compute.sigmoid(-100)
```

```python
x = list(range(-11,11))
y = [Activations.sigmoid(i) for i in x]
plt.plot(x,y)
plt.title("sigmoid")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
x = list(range(-11,11))
y = [Activations.tanh(i) for i in x]
plt.plot(x,y)
plt.title("tanh")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
x = list(range(-11,11))
y = [Activations.ReLU(i) for i in x]
plt.plot(x,y)
plt.title("sigmoid")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
x = list(range(-11,11))
y = [Activations.leakyReLU(i) for i in x]
plt.plot(x,y)
plt.title("sigmoid")
plt.xlabel("X")
plt.ylabel("Y")
```

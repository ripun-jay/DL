### Import required packages

```python
import tensorflow as tf
import pandas as pd
import numpy as np
from tensorflow import keras
```

```python
from matplotlib import pyplot as plt
import matplotlib
```

#### Loading Dataset

```python
fashion = keras.datasets.fashion_mnist.load_data()
```

#### Train test split

```python
(X_train, y_train), (X_test, y_test) = fashion
```

##### Looking at images


```python
classes = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]

def show_image(index):   #pass index to see images
        plt.imshow(X_train[index], cmap= matplotlib.cm.binary, interpolation = "hanning")
        # plt.title(classes[y_train[index]])
        plt.title(classes[y_train[index]])
        plt.axis(False)
    
show_image(1)
```

```python
show_image(90)
```

##### Scaling training data for better performace

```python
X_train_scaled = X_train/255
X_train_scaled.shape
```

###### converting categorical label to OneHot Encoded data 

```python
# y_train_categorical= keras.utils.to_categorical(y_train, num_classes= 10, dtype= "float32")
# y_train_categorical[:5]
```

#### Creating Artificial Neural Network for Image classification
* with TesorBoard initialization

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

tb_callback = tf.keras.callbacks.TensorBoard(log_dir="logs/", histogram_freq=1, profile_batch = 100000000)
```

##### train this ANN

```python
model.fit(X_train_scaled, y_train, epochs=500 , callbacks=[tb_callback],)
```

```python
y_pred = model.predict(X_train_scaled)
```
